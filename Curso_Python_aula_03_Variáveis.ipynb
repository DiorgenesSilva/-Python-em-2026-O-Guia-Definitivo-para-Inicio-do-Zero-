```python
# Variável adicionando Texto (String )
nome_do_canal = "Programador Iniciante"
print(nome_do_canal)

# Número Interiro (Int)
videos_postados = 42
print(videos_postados)

# Numero decimal (Float ou Ponto Flutuante)
versao_python = 3.12
print(versao_python)

# Verdadeiro ou Falso (Booleano)
curso_concluido = True
print(curso_concluido)

print('O meu canal',nome_do_canal, 'nele eu tenho',videos_postados,'no total')
print('Estou utilizando a versao',versao_python,'do Python')
print('E meu curso esta como',curso_concluido, 'Finalmente')
```
